```python
import pandas as pd
from google.colab import files
uploaded = files.upload()

data = pd.read_csv("data.csv", error_bad_lines=False)
print(data.head())
#data=pd.read_csv('/kaggle/input/password-strength-classifier/data.csv',on_bad_lines='skip')
print(data.columns)
print(data.describe())
print(data.info())
print(data.isna().sum())
data['password']=data['password'].fillna(data['password'].mode())
print(data.isna().sum())
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
features=tfidf.fit_transform(data['password'].head(10000))
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,data['strength'].head(10000))
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
print(rf.score(x_test,y_test))
from sklearn.tree import DecisionTreeClassifier
tre=DecisionTreeClassifier()
tre.fit(x_train,y_train)
print(tre.score(x_test,y_test))
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
print(xgb.score(x_test,y_test))
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=200,multi_class='multinomial')
lr.fit(x_train,y_train)
print(lr.score(x_test,y_test))
pred=lr.predict(tfidf.transform([input('enter the password: ')]))
pred1=rf.predict(tfidf.transform([input('enter the password: ')]))
pred2=xgb.predict(tfidf.transform([input('enter the password: ')]))
pred3=tre.predict(tfidf.transform([input('enter the password: ')]))
if pred == 1 or pred1==1 or pred2==1 or pred3==1:
    print('moderate')
elif pred == 0 or pred1==0 or pred2==0 or pred3==0:
    print('password is weak')
elif pred == 2 or pred1==2 or pred2==2 or pred3==2:
    print('password is strong')
```
